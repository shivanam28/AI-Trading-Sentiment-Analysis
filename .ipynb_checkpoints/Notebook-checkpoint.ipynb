```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

trades = pd.read_csv("historical_data.csv")
sentiment = pd.read_csv("fear_greed_index.csv")
```

```python
print(trades.head())
print(trades.columns)

print(sentiment.head())
print(sentiment.columns)
```

```python
print(trades.isnull().sum())
print(sentiment.isnull().sum())
```

```python
trades['Timestamp IST'] = pd.to_datetime(trades['Timestamp IST'], format='%d-%m-%Y %H:%M')
trades['date'] = trades['Timestamp IST'].dt.date
sentiment['date'] = pd.to_datetime(sentiment['date']).dt.date
```

```python
merged = pd.merge(
    trades,
    sentiment,
    on='date',
    how='left'
)
```

```python
print(merged.head())
print(merged[['date', 'classification']].head())
```

```python
merged['profit_flag'] = merged['Closed PnL'] > 0
```

```python
print("Total Trades:", len(merged))
print("Total Traders:", merged['Account'].nunique())
print("Total Coins:", merged['Coin'].nunique())
```

```python
plt.figure(figsize=(8,5))
sns.countplot(data=merged, x='classification')
plt.title("Market Sentiment Distribution")
plt.show()
```

```python
avg_pnl = merged.groupby('classification')['Closed PnL'].mean()

print(avg_pnl)
```

```python
print(merged['Direction'].unique())
```

```python
plt.figure(figsize=(8,5))
avg_pnl.plot(kind='bar')
plt.title("Average Closed PnL by Sentiment")
plt.ylabel("Average Profit/Loss")
plt.show()
```

```python
volume = merged.groupby('classification')['Size USD'].mean()

print(volume)
```

```python
plt.figure(figsize=(8,5))

volume.plot(kind='bar')

plt.title("Average Trade Size by Sentiment")

plt.ylabel("Trade Size USD")

plt.show()
```

```python
side_analysis = pd.crosstab(
    merged['classification'],
    merged['Side']
)

print(side_analysis)
```

```python
side_analysis.plot(
    kind='bar',
    stacked=True,
    figsize=(8,5)
)

plt.title("Buy/Sell Behavior vs Sentiment")

plt.show()
```

```python
win_rate = merged.groupby(
    'classification'
)['profit_flag'].mean() * 100

print(win_rate)
```

```python
plt.figure(figsize=(8,5))

win_rate.plot(kind='bar')

plt.title("Win Rate by Sentiment")

plt.ylabel("Win Rate %")

plt.show()
```

```python
top_coins = merged.groupby(
    'Coin'
)['Closed PnL'].sum().sort_values(ascending=False).head(10)

print(top_coins)
```

```python
plt.figure(figsize=(10,6))

top_coins.plot(kind='bar')

plt.title("Top Profitable Coins")

plt.ylabel("Total Profit")

plt.show()
```

```python
plt.figure(figsize=(10,6))

sns.boxplot(
    data=merged,
    x='classification',
    y='Closed PnL'
)

plt.title("Profit Distribution by Sentiment")

plt.show()
```

```python
import numpy as np
numeric_cols = merged.select_dtypes(include=np.number)

plt.figure(figsize=(12,8))

sns.heatmap(
    numeric_cols.corr(),
    annot=True,
    cmap='coolwarm'
)

plt.title("Correlation Heatmap")

plt.show()
```

```python
merged.to_csv("merged_data.csv", index=False)
```
